# Closing price analysis

```python
import yfinance as yf
import pandas as pd
import numpy as np

def get_closing_prices(stock_symbol, year):
    start_date = f"{year}-01-01"
    end_date = f"{year}-12-31"


    stock_data = yf.download(stock_symbol, start=start_date, end=end_date)


    stock_data.sort_index(inplace=True)
    closing_prices = stock_data['Close'].tolist()

    return closing_prices


def get_average(stock_symbol,year=[]):
  samp=[]
  if(year==[]):
    for i in range(2008,2025):
      closing_prices = get_closing_prices(stock_symbol, i)
      samp.append(np.sum(np.array(closing_prices))/len(list(closing_prices)))
  else:
    for i in year:
      closing_prices = get_closing_prices(stock_symbol, i)
      samp.append(np.sum(np.array(closing_prices))/len(list(closing_prices)))

  print("")
  print(samp)
  return samp

```

```python
def extract_st(text):
    pattern =  r'\|.*?\]'
    samp= list(re.findall(pattern, text))
    ans=[]
    for i in samp:
      ans.append(i[1:-1])

    return ans
```

```python
import seaborn as sns
```

```python
def create_heatmap_from_dict(data_dict, label):
    # Extract all unique sources and targets
    sources = set()
    targets = set()
    for (source, target) in data_dict.keys():
        sources.add(source)
        targets.add(target)

    sources = sorted(sources)  # Sort for consistent ordering
    targets = sorted(targets)  # Sort for consistent ordering

    # Initialize a DataFrame with zeros
    heatmap_data = pd.DataFrame(0, index=targets, columns=sources)

    # Fill the DataFrame with the counts from the dictionary
    for (source, target), count in data_dict.items():
        heatmap_data.at[target, source] = count

    # Create a mask for zero values
    mask = heatmap_data == 0

    # Create a custom colormap
    cmap = sns.color_palette("coolwarm", as_cmap=True)
    cmap.set_bad(color='grey')  # Set color for NaNs (which we will use for zeros)

    # Convert zeros to NaNs for masking purposes
    heatmap_data = heatmap_data.astype(float)
    heatmap_data[mask] = np.nan

    # Plotting the heatmap
    plt.figure(figsize=(5, 5))
    sns.heatmap(heatmap_data, annot=False, cmap=cmap, cbar_kws={'label': 'Count'}, linewidths=0.5, linecolor='grey')
    plt.xlabel('Source Tags')
    plt.ylabel('Target Tags')
    plt.title(label)
    plt.show()
```

```python
def sort_dict(word_counts):  #to get top 10 mappings for plotting
  sorted_word_counts_desc = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
  temp={}
  for i in sorted_word_counts_desc[:15]:
    temp[i[0]]=i[1]

  return temp
```

```python
def combine_dictionaries(dicts):
    # print(dicts)
    combined_dict = {}
    for d in dicts:
        for key, value in d.items():
            if key in combined_dict:
                combined_dict[key] += value
            else:
                combined_dict[key] = value
    return combined_dict
```

### Concept Mapping after a threshold barrier


##### Criteria used for finding representative keys and removing the overlapping keys

For each dictionary we first find the representative keys by the following method:
1. Normalize the values by the maximum of each dict.
2. Rank keys by the normalised values from high to low.
3. If the value of a key in dict_1 belongs the first quartile and do not belong to the first quartiles of other dicts,. We find 10 such representative keys.
4. Now for all the common keys we check if the key is representative key of any dictionary (say it is for dict1 and dict2) then we do not affect this key in that dictionary, for the other dictionaries (other than dict1 and dict2) we check if the value of this key is greater than some "threshold" value. If it is then we delete this key from that dictionary else, we let it be in the other dictionary. Threshold is chosen such  that in the normalized values of keys (say we are checking if the common key which is not a representative key of dict1, should remain in dict1 or not, so we will take the normalized values of keys of dict1) we chose the second lowest value and treat that as threshold for checking that dictionary (dict1)

```python
import numpy as np

def normalize_dict(d):
    max_value = max(d.values())
    return {k: v / max_value for k, v in d.items()}

def rank_keys_by_value(d):
    return sorted(d, key=d.get, reverse=True)

def get_first_quartile_keys(ranked_keys, n):
    return ranked_keys[:n // 4]

def find_representative_keys(dicts):
    normalized_dicts = [normalize_dict(d) for d in dicts]
    ranked_keys = [rank_keys_by_value(d) for d in normalized_dicts]
    n = min(len(keys) for keys in ranked_keys)
    first_quartile_keys = [set(get_first_quartile_keys(rk, n)) for rk in ranked_keys]

    representative_keys = []
    for i in range(len(dicts)):
        rep_keys = first_quartile_keys[i] - set.union(*[first_quartile_keys[j] for j in range(len(dicts)) if j != i])
        representative_keys.append(set(list(rep_keys)[:10]))

    return representative_keys

def modify_dicts(dicts):
    representative_keys = find_representative_keys(dicts)
    normalized_dicts = [normalize_dict(d) for d in dicts]

    all_keys = set.union(*[set(d.keys()) for d in dicts])
    common_keys = set.intersection(*[set(d.keys()) for d in dicts])

    for key in common_keys:
        rep_in_dicts = [i for i, reps in enumerate(representative_keys) if key in reps]

        if not rep_in_dicts:
            thresholds = [sorted(d.values())[0] for d in normalized_dicts]
            for i, d in enumerate(dicts):
                if i not in rep_in_dicts and normalized_dicts[i].get(key, 0) > thresholds[i]:
                    del d[key]

    return dicts
```
